```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
import time
import datetime
from datetime import timedelta
```

```python
ID_141=pd.read_csv('C:\\Users\\qiu\\Downloads\\Daten\\TagID_141_2019.csv',dtype=str,parse_dates=['CreateDateTime'],index_col='Unnamed: 0')
```

```python
#ID_141['DeltaTime']=ID_141['CreateDateTime'].diff
```

```python
Daten=ID_141.CreateDateTime.diff(2).dt.total_seconds()
```

```python
ID_141['DeltaTime']=Daten
```

```python
ID_141
```

```python
ID_141['Frequenz']=1/ID_141['DeltaTime']
```

```python
ID_141.dtypes
```

```python
ID_141.head(100)
```

```python
ID_141['Frequenz'].fillna(value='0',inplace=True)
```

```python
ID_141['Frequenz']=ID_141['Frequenz'].astype(float)
```

```python
ID_141.plot(x='CreateDateTime',y='Frequenz',figsize=(18, 6),title='ID_141 in Hz')
```
